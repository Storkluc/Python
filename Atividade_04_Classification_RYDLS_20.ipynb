```python
import numpy as np
from skimage.feature import local_binary_pattern
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn import  model_selection


from sklearn.metrics import make_scorer, accuracy_score, precision_score, f1_score, recall_score
from sklearn.model_selection import train_test_split, StratifiedKFold
import warnings

warnings.filterwarnings('ignore')
```

```python
dataset_pulmoes = pd.read_csv('/content/feature_matrix_train.csv')
```

```python
#Carregando a base e separando em treinamento e teste
dataset_pulmoes = pd.read_csv('/content/feature_matrix_train.csv')
y = dataset_pulmoes.iloc[:,-1:]
y = pd.array(y['class'])
dataset_pulmoes.drop(axis=1, columns='class', inplace=True)
X = dataset_pulmoes

#Separando a base em treinamento e teste estratificado
skfold = StratifiedKFold(n_splits=5,shuffle=True, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.3, random_state=42, stratify=y)
```

```python
#Classificadores
knn = KNeighborsClassifier(n_neighbors=2, weights='distance', algorithm='auto')
dt = tree.DecisionTreeClassifier(criterion='gini', min_samples_split=2,splitter='best',random_state=8)
nb = GaussianNB(var_smoothing=0.2)
lr = LogisticRegression(random_state=42, solver='lbfgs', class_weight='balanced', warm_start=True, intercept_scaling=0.3)
rf = RandomForestClassifier(n_estimators=100, criterion='gini',random_state=8,n_jobs=8)
ab = AdaBoostClassifier(n_estimators=50, random_state=24, learning_rate=1.3)
```

```python
#Treinando os modelos
clknn = knn.fit(X_train, y_train)
cldt = dt.fit(X_train, y_train)
clnb = nb.fit(X_train, y_train)
cllr = lr.fit(X_train, y_train)
clrf = rf.fit(X_train, y_train)
clab = ab.fit(X_train, y_train)
```

```python
# Teste usando a base de testes
y_pred_clknn = clknn.predict(X_test)
y_pred_cldt = cldt.predict(X_test)
y_pred_clnb = clnb.predict(X_test)
y_pred_cllr = cllr.predict(X_test)
y_pred_clrf = clrf.predict(X_test)
y_pred_clab = clab.predict(X_test)
```

```python
def metricas(basetest, basepred):
  metricas = []
  metricas.append(accuracy_score(basetest, basepred))
  metricas.append(precision_score(basetest, basepred, average='macro'))
  metricas.append(f1_score(basetest, basepred, average='macro'))
  metricas.append(recall_score(basetest, basepred, average='macro'))
  return metricas
```

```python
#Gerando as matrizes de confusão
matriz_clknn = pd.crosstab(y_test,y_pred_clknn,rownames=['Real'], colnames=['Previsto'], margins=True)
matriz_cldt = pd.crosstab(y_test,y_pred_cldt,rownames=['Real'], colnames=['Previsto'], margins=True)
matriz_clnb = pd.crosstab(y_test,y_pred_clnb,rownames=['Real'], colnames=['Previsto'], margins=True)
matriz_cllr = pd.crosstab(y_test,y_pred_cllr,rownames=['Real'], colnames=['Previsto'], margins=True)
matriz_clrf = pd.crosstab(y_test,y_pred_clrf,rownames=['Real'], colnames=['Previsto'], margins=True)
matriz_clab = pd.crosstab(y_test,y_pred_clab,rownames=['Real'], colnames=['Previsto'], margins=True)
```
